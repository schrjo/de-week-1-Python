````python
#1. **Middle char(s)**
#* `middle(s)` → If odd length return the middle char, else return the two middle chars.

def middle(s):
    indexpos = int(len(s)/2)

    if len(s) % 2 == 0:
        return s[indexpos - 1: indexpos + 1]
    else:
        return s[indexpos]


assert middle("abc") == "b"
assert middle("abba") == "bb"
````
